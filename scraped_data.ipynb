```python
import snscrape.modules.twitter as sns
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# scraping tweets related to  differet retail compenies in india over a period 2019 - 2021
```

```python
tweets = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sns.TwitterSearchScraper('#ajio since:2019-01-01 until:2021-12-25').get_items()):
    if i>5000:
        break
    tweets.append([tweet.date, tweet.id,tweet.replyCount,tweet.likeCount,tweet.renderedContent])
    
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets, columns=['Datetime', 'Tweet Id','replycount','like','content'])
```
